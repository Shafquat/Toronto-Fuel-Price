```python
import pandas as pd
import numpy as np

%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# conda install -c conda-forge ffmpeg
from IPython.display import HTML
```

```python
gas_prices = pd.read_csv('D:/Dropbox/Data/gas-prices-Toronto/avg-gas-price.csv')
```

```python
x = np.array(gas_prices.iloc[:,0])
y = gas_prices.iloc[:,1]

fig, ax = plt.subplots(figsize=(22,8))
plt.xticks(rotation=45)
line, = ax.plot(x, y, color='purple')
ax.set_xticks(x[::6])
ax.set_xlabel("Month-Year")
ax.set_ylabel("Cents/Litre")
ax.set_title("Avg. Cost of Regular Unleaded Gasoline in Toronto")

def update(num, x, y, line):
    line.set_data(x[:num], y[:num])
    line.axes.axis([0, 363, 0, 150])
    return line,
```

```python
# call the animator. blit=True means only re-draw the parts that have changed.
anim = animation.FuncAnimation(fig, update, len(x), fargs=[x, y, line],
                              interval=35, blit=True, repeat=False)
```

```python
HTML(anim.to_html5_video())
```
